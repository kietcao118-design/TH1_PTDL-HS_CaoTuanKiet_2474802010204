```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dulieuxettuyendaihoc.csv")
df.head()
```

```python
dinh_tinh = ['GT', 'DT', 'KV', 'KT']

dinh_luong = [col for col in df.columns if col not in dinh_tinh]

print("Định tính:")
print(dinh_tinh)

print("\nĐịnh lượng:")
print(dinh_luong)
```

```python
print(df.head(10))

print(df.tail(10))
```

```python
print(df['DT'].isnull().sum())

print(df['DT'].unique())

print(df['DT'].value_counts(dropna=False))

df['DT'] = df['DT'].fillna(0)
```

```python
print(df['T1'].isnull().sum())

mean_T1 = df['T1'].mean()

df['T1'] = df['T1'].fillna(mean_T1)
```

```python
cot_diem = [
    'T1','L1','H1','S1','V1','X1','D1','N1',
    'T2','L2','H2','S2','V2','X2','D2','N2',
    'T6','L6','H6','S6','V6','X6','D6','N6',
    'DH1','DH2','DH3'
]

for col in cot_diem:

    mean_value = df[col].mean()

    df[col] = df[col].fillna(mean_value)
```

```python
df['TBM1'] = (
    df['T1']*2 + df['L1'] + df['H1'] + df['S1']
    + df['V1']*2 + df['X1'] + df['D1'] + df['N1']
) / 10

df['TBM2'] = (
    df['T2']*2 + df['L2'] + df['H2'] + df['S2']
    + df['V2']*2 + df['X2'] + df['D2'] + df['N2']
) / 10

df['TBM3'] = (
    df['T6']*2 + df['L6'] + df['H6'] + df['S6']
    + df['V6']*2 + df['X6'] + df['D6'] + df['N6']
) / 10
```

```python
def xep_loai(tbm):

    if tbm < 5:
        return 'Y'

    elif tbm < 6.5:
        return 'TB'

    elif tbm < 8:
        return 'K'

    elif tbm < 9:
        return 'G'

    else:
        return 'XS'
```

```python
df['XL1'] = df['TBM1'].apply(xep_loai)

df['XL2'] = df['TBM2'].apply(xep_loai)

df['XL3'] = df['TBM3'].apply(xep_loai)
```

```python
def min_max(series):

    return ((series - series.min()) /
           (series.max() - series.min())) * 4

df['US_TBM1'] = min_max(df['TBM1'])

df['US_TBM2'] = min_max(df['TBM2'])

df['US_TBM3'] = min_max(df['TBM3'])
```

```python
def ket_qua(row):

    khoi = row['KT']

    if khoi in ['A', 'A1']:

        diem = (row['DH1']*2 + row['DH2'] + row['DH3']) / 4

    elif khoi == 'B':

        diem = (row['DH1'] + row['DH2']*2 + row['DH3']) / 4

    else:

        diem = (row['DH1'] + row['DH2'] + row['DH3']) / 3

    if diem >= 5:
        return 1
    else:
        return 0

df['KQXT'] = df.apply(ket_qua, axis=1)
```

```python
df.to_csv("processed_dulieuxettuyendaihoc.csv", index=False)

print("Đã lưu file")
```
