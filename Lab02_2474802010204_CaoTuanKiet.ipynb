```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot
```

```python
# Đọc dữ liệu từ file CSV
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")
```

Phần 1: Thống Kê Dữ Liệu

```python
# CÂU 1 — SẮP XẾP DH1 TĂNG DẦN
df.sort_values("DH1").head()
```

```python
# CÂU 2 
df.sort_values(["GT", "DH2"]).head()
# Ham pivot dung chung
agg = [
    "count","sum","mean","median","min","max","std",
    lambda x: x.quantile(0.25),
    lambda x: x.quantile(0.50),
    lambda x: x.quantile(0.75)
]
```

```python
#Câu 3
pivot_KT = df.pivot_table(
    values="DH1",
    index="KT",
    aggfunc=agg
)

print(pivot_KT)
```

```python
# CÂU 4 
pivot_KT_KV = df.pivot_table(
    values="DH1",
    index=["KT","KV"],
    aggfunc=agg
)

print(pivot_KT_KV)
```

```python
# CÂU 5 
pivot_KT_KV_DT = df.pivot_table(
    values="DH1",
    index=["KT","KV","DT"],
    aggfunc=agg
)

print(pivot_KT_KV_DT)
```

```python
# Câu 2
df_loc = df[
    (df["KT"].isin(["A","A1","B"])) &
    (df["KV"].isin(["1","2"]))
]

pd.crosstab(
    [df_loc["KT"], df_loc["KV"]],
    df_loc["KQXT"]
).plot(kind="bar")

plt.title("KQXT theo KT va KV")
plt.show()
```

```python
# Câu 6
pd.crosstab(df["DT"], df["KQXT"]).plot(kind="bar")

plt.title("Dau rot theo DT")
plt.show()
```

#Phần 4: Trực quan hóa dữ liệu nâng cao

```python
#CÂU 1 
# Đếm tần số từng giá trị T1
tan_so_T1 = df["T1"].value_counts().sort_index()

print(tan_so_T1)

# Vẽ biểu đồ đường
plt.plot(
    tan_so_T1.index,
    tan_so_T1.values
)

plt.title("Bieu do duong Simple cho T1")
plt.xlabel("T1")
plt.ylabel("Count")

plt.xticks(rotation=90)

plt.show()
```

```python
# Câu 2
def phanloai(x):

    if x < 5:
        return "k"

    elif x < 7:
        return "tb"

    elif x < 8:
        return "kha"

    return "g"

df["phanlopt1"] = df["T1"].apply(phanloai)
```

```python
# Câu 3
print(df["phanlopt1"].value_counts())
```

```python
# CÂU 4 
bang = pd.crosstab(df["T1"], df["phanlopt1"])
print(bang)
# Vẽ multiple line
bang.plot(kind="line", marker="o")
plt.title("Multiple Line cua T1 theo phanlopt1")
plt.xlabel("T1")
plt.ylabel("Count")

plt.xticks(rotation=90)

plt.legend(title="PhanlopT1")

plt.show()
```

```python
# CÂU 5 
# Đếm tần số
bang = pd.crosstab(df["T1"], df["phanlopt1"])

# Vẽ từng nhóm
for cot in bang.columns:

    markerline, stemlines, baseline = plt.stem(
        bang.index,
        bang[cot],
        label=cot
    )

plt.title("Drop-line cua T1 theo phanlopt1")
plt.xlabel("T1")
plt.ylabel("Count")

plt.xticks(rotation=90)

plt.legend(title="PhanlopT1")

plt.show()
```

#Phần 5:Mô tả dữ liệu và khảo sát dạng phân phối

```python
# Cau 1
print(df["T1"].describe())

print("Variance:", df["T1"].var())
print("Skewness:", df["T1"].skew())
print("Kurtosis:", df["T1"].kurt())

plt.boxplot(df["T1"])
plt.title("Boxplot T1")
plt.show()

plt.hist(df["T1"])
plt.title("Histogram T1")
plt.show()

qqplot(df["T1"], line="s")
plt.title("QQPlot T1")
plt.show()
```

```python
#Câu 2
for ten, data in df.groupby("phanlopt1"):

    plt.boxplot(data["T1"])
    plt.title(f"Boxplot {ten}")
    plt.show()

    plt.hist(data["T1"])
    plt.title(f"Histogram {ten}")
    plt.show()
```

```python
# Câu 3
print(df[["DH1","T1"]].corr())

plt.scatter(df["T1"], df["DH1"])

plt.xlabel("T1")
plt.ylabel("DH1")
plt.show()
```

```python
# Câu 4
for kv, data in df.groupby("KV"):

    print(data[["DH1","T1"]].corr())

    plt.scatter(data["T1"], data["DH1"])

    plt.title(f"KV {kv}")
    plt.show()
```

```python
# Câu 5
print(df[["DH1","DH2","DH3"]].corr())

plt.scatter(df["DH1"], df["DH2"])
plt.show()

plt.scatter(df["DH1"], df["DH3"])
plt.show()

plt.scatter(df["DH2"], df["DH3"])
plt.show()
```
